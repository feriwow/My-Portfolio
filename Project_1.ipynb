# Bike-share analysis case study

## Overview
The purpose of this project is to give insight of Bike-share activity at Cyclistic, A bike-share program that features more than 5,800 bicycles and 600 docking stations. The goal of this analysis is to found out casual riders and annual members use Cyclistic bikes differently and invent new strategy to convert casual riders into annual members. This dataset contain data from the customer of Cyclistic in span of one month. 


## Business Understanding
In 2016, Cyclistic launched a successful bike-share offering. The bikes can be unlocked from one station and returned to any other station in the system anytime.
Until now, Cyclistic’s marketing strategy relied on building general awareness and appealing to broad consumer segments. One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes, and annual memberships. Customers who purchase single-ride or full-day passes are referred to as casual riders. Customers who purchase annual memberships are Cyclistic members.

To formulate the analysis formulation there are couple of aspects needed to be covered:
- The Problem   : The marketing department wants to build a better campaign that could convert casual rider to annual member. This will be the key for the future growth of the company. since annual members are much more profitable than casual riders. 
- Data          : The data sourced sourced from https://divvy-tripdata.s3.amazonaws.com/index.html
- Objective     : Find the determinant variables and how annual members and casual riders use bikes differently so that it could help Casual rider to convert to annual member.
- Action        : Create a summary of the analysis and give recommendations based on the analysis.
- Value         :  This analysiswill help the company to make more profitsince annual members are much more profitable than casual riders. 

```python
import numpy as np 
import pandas as pd 
```

```python
activity = pd.read_csv('tripdata_2.csv')
activity
```

```python
activity = activity.dropna()
activity2 = activity.copy()
```

```python
activity2['started_at'] = activity2['started_at'].astype(np.datetime64)
activity2['ended_at'] = activity2['ended_at'].astype(np.datetime64)
activity2.info()
```

```python
activity2['ride_length(minute)'] = (activity2['ended_at'] - activity2['started_at']).astype('timedelta64[m]')
```

```python
cols = activity2.columns.tolist()
cols
```

```python
activity2 = activity2[['ride_id',
 'rideable_type',
 'started_at',
 'ended_at',
 'ride_length(minute)',
 'start_station_name',
 'start_station_id',
 'end_station_name',
 'end_station_id',
 'start_lat',
 'start_lng',
 'end_lat',
 'end_lng',
 'member_casual']]
```

```python
activity2['day_of_week'] = activity2['started_at'].dt.day_name()
activity2
```

```python
activity2['day_of_week'].mode()
```

```python
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
activity2['day_of_week'] = pd.Categorical(activity2['day_of_week'], categories=cats, ordered=True)
activity2 = activity2.sort_values('day_of_week')
```

```python
activity2.describe()
```

```python
means = activity2['ride_length(minute)'].groupby(activity2['member_casual']).mean()
means
```

```python
day = activity2['ride_length(minute)'].groupby([activity2['member_casual'], activity2['day_of_week']]).mean()
day
```

```python
day2 = activity2['ride_length(minute)'].groupby([activity2['member_casual'], activity2['day_of_week']])
day2.count()
```

```python
import plotly.express as px
fig = px.histogram(
    activity2, 
    x='day_of_week', 
    color='member_casual',
    opacity= 0.5
    )
fig.show()
```

Based on the chart above, it can be seen that for annual members, wednesday is the most active with a total of 5793 riders per day on avarage. While the casual members are the most active on saturday with 3122 riders per day on avarage. overall, the day with the most bikes rent by rider happen on saturday, this probably due to saturday being a weekend so many people choose cycling as their holiday activity. 


Casual riders most active during this day probably due to holiday and casual riders only have time for cycling during their day off. While annual members mostly active during the weekday, this is probably because annual members use bicycle as a mean  of transportaion to their workplace.

```python
d5 = day.to_frame()
d6 = d5.reset_index()
d6.sort_values(by = 'ride_length(minute)')
```

```python
fig = px.bar(
    d6, 
    x='day_of_week',
    y = 'ride_length(minute)',
    color = 'member_casual',
    color_discrete_map= {'member' : 'blue',
    'casual': 'red'},
    opacity= 0.5
    )
fig.show()
```

Based on the chart above, it can be seen that on avarage, casual riders spent more time on the ride than annual members even on a weekday. This probably caused by the travel speed. Because casual riders who do not cycling regulary is slower than anual members who cycling frequently. For casual biker the ride length is peak on friday at avarage of 61 minute.
While for member biker is peak at sunday where they can cycling at ease because its weekend at avarage of 16 minutes.
